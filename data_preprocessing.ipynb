# 2 Preprocessing and Cleaning

```python
# Change the wide DataFrame to long DataFrame by given format.
def melt(wide_df: pd.DataFrame, col_list: list, val_name: str) -> pd.DataFrame: 
    long_df =  wide_df.melt(id_vars = col_list[:2], value_vars = col_list[2:], var_name = "Date", value_name = val_name)
    long_df["Date"] = pd.to_datetime(long_df["Date"])
    return long_df
```

```python
col_list = confirmed_df.columns.values.tolist()

long_confirmed_df = melt(confirmed_df, col_list, "Confirmed")
long_deaths_df = melt(deaths_df, col_list, "Deaths")
long_recovered_df = melt(recovered_df, col_list, "Recovered")
```

```python
long_deaths_df.sample(10)
```

```python
long_recovered_df.sample(10)
```

```python
# Merge the three long DataFrames.
col_list = col_list[:4].append("Date")

merged_df = pd.merge(left = long_confirmed_df, right = long_deaths_df, how = "left", on = col_list)
merged_df = pd.merge(left = merged_df, right = long_recovered_df, how = "left", on = col_list)

merged_df.sample(10)
```

```python
# pd.set_option('display.max_rows', None)
# display(merged_df["Country/Region"].value_counts().to_frame())
```

```python
# Add Active cases.
merged_df["Active"] = merged_df["Confirmed"] - merged_df["Deaths"] - merged_df["Recovered"]
```

```python
# Fill NaN with integer 0 for four status.
status = merged_df.iloc[:, -4:].columns.values.tolist()
merged_df[status] = merged_df[status].fillna(0)
merged_df[status] = merged_df[status].astype("int")
```

```python
# Fill NaN with None for Province/State column.
merged_df["Province/State"] = merged_df["Province/State"].fillna(" ")
```

```python
# Make all the country's name with the same format.
merged_df["Country/Region"] = merged_df["Country/Region"].replace("Korea, North", "North Korea")
merged_df["Country/Region"] = merged_df["Country/Region"].replace("Korea, South", "South Korea")
merged_df["Country/Region"] = merged_df["Country/Region"].replace("US", "United States")
```

```python
def add_country_or_region(WHO: dict, name: str, countries_or_regions: str) -> dict:
    country_or_region = [i.strip() for i in countries_or_regions.split(",")]
    for i in country_or_region:
        WHO[i] = name
    return WHO

processed_WHO_regions = {}
for key, value in WHO_regions.items():
    processed_WHO_regions = add_country_or_region(processed_WHO_regions, key, value)

processed_WHO_regions
```

```python
merged_df["WHO Region"] = merged_df["Country/Region"].map(processed_WHO_regions)
```
