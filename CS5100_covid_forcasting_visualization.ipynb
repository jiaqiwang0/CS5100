```python
cleaned_data = pd.read_csv(path+'/covid_19_clean_complete.csv', parse_dates=['Date'])

cases = ['Confirmed', 'Deaths', 'Recovered', 'Active']

# active case = confirmed - deaths - recovered
cleaned_data['Active'] = cleaned_data['Confirmed'] - cleaned_data['Deaths'] - cleaned_data['Recovered']

cleaned_data[['Province/State']] = cleaned_data[['Province/State']].fillna('')
cleaned_data[cases] = cleaned_data[cases].fillna(0)

cleaned_data[:]
```

# **Analysis and Visualization**

## 1. Various Type of Cases Worldwide
According to the data set we have, we can visualize the total number of various types of cases worldwide. In this part, we will mainly include four types of cases, which are confirmed cases, death cases, recovered cases and active cases. By delivering these outcomes, we could have a better vision of how Covid cases look like over the world and prepare for the following data learning and analyzing.

### 1.1 Confirmed Cases Worldwide

```python
# confirmed cases worldwide
confirmied_cases = pd.DataFrame(cleaned_data.groupby('Country/Region')['Confirmed'].sum())
confirmied_cases['Country/Region'] = confirmied_cases.index
confirmied_cases.index = np.arange(1,188)

global_confirmied_cases = confirmied_cases[['Country/Region','Confirmed']]

fig = px.bar(global_confirmied_cases.sort_values('Confirmed',ascending=False)[:20][::-1],x='Confirmed',y='Country/Region',title='Confirmed Cases Worldwide',text='Confirmed', height=900, orientation='h')
fig.show()
```

### 1.2 Deaths Cases Worldwide

```python
# death cases worldwide
death_cases = pd.DataFrame(cleaned_data.groupby('Country/Region')['Deaths'].sum())
death_cases['Country/Region'] = death_cases.index

global_death_cases = death_cases[['Country/Region','Deaths']]

fig = px.bar(global_death_cases.sort_values('Deaths',ascending=False)[:20][::-1],x='Deaths',y='Country/Region',title='Deaths Cases Worldwide',text='Deaths', height=900, orientation='h')
fig.show()
```

### 1.3 Recovered Cases Worldwide

```python
# recovered cases worldwide
recovered_cases = pd.DataFrame(cleaned_data.groupby('Country/Region')['Recovered'].sum())
recovered_cases['Country/Region'] = recovered_cases.index

global_recovered_cases = recovered_cases[['Country/Region','Recovered']]

fig = px.bar(global_recovered_cases.sort_values('Recovered',ascending=False)[:20][::-1],x='Recovered',y='Country/Region',title='Recovered Cases Worldwide',text='Recovered', height=900, orientation='h')
fig.show()
```

### 1.4 Active Cases Worldwide

```python
# active cases worldwide
active_cases = pd.DataFrame(cleaned_data.groupby('Country/Region')['Active'].sum())
active_cases['Country/Region'] = active_cases.index

global_active_cases = active_cases[['Country/Region','Active']]

fig = px.bar(global_active_cases.sort_values('Active',ascending=False)[:20][::-1],x='Active',y='Country/Region',title='Active Cases Worldwide',text='Active', height=900, orientation='h')
fig.show()
```

> By the bar graphs, we can understand that:
1. US, Brazil, Russia and India are in both of top 4 places global confirmed cases and recovered cases
2. US, United Kindom, Brazil and Italy are in top 4 places of global death cases
3. US, Brazil, United Kingdom and Russia are in top 4 places of global active cases

## 2. Spread of Coronavirus over Time Globally and Regionally
To display the spread of Covid over time, we will utilize scatter plots, where the x-axis represents the time period and the y-axis represents the total number of cases. In this part, we will show the spread of Covid over time in the following regions: Worldwide, China, Italy, Iran, South Korea, Spain and the rest of the world. For each graph, we will also include four categories of cases, which are confirmed, death, recovered and active. 

2.1 Global Spread of the Covid-19 Over Time

```python
date_1 = cleaned_data.groupby('Date')['Confirmed'].sum().reset_index()
date_2 = cleaned_data.groupby('Date')['Deaths'].sum().reset_index()
date_3 = cleaned_data.groupby('Date')['Recovered'].sum().reset_index()
date_4 = cleaned_data.groupby('Date')['Active'].sum().reset_index()

fig = make_subplots(rows=1, cols=4, subplot_titles=("Comfirmed", "Deaths", "Recovered",'Active'))

trace1 = go.Scatter(
                x=date_1['Date'],
                y=date_1['Confirmed'],
                name="Confirmed",
                mode='lines',
                line_color='orange',
                opacity=0.8)

trace2 = go.Scatter(
                x=date_2['Date'],
                y=date_2['Deaths'],
                name="Deaths",
                mode='lines',
                line_color='red',
                opacity=0.8)

trace3 = go.Scatter(
                x=date_3['Date'],
                y=date_3['Recovered'],
                name="Recovered",
                mode='lines',
                line_color='green',
                opacity=0.8)

trace4 = go.Scatter(
                x=date_4['Date'],
                y=date_4['Active'],
                name="Active",
                mode='lines',
                line_color='blue',
                opacity=0.8)

fig.append_trace(trace1, 1, 1)
fig.append_trace(trace2, 1, 2)
fig.append_trace(trace3, 1, 3)
fig.append_trace(trace4, 1, 4)
fig.update_layout(template="plotly_dark",title_text = '<b>Global Spread of the Coronavirus Over Time </b>', font=dict(color='white'))
fig.show()
```
